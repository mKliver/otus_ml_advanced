### 1 Путешествие в страну Х
#### Задача:
Вы отправляетесь в страну X, которая славится своими мистическими приключениями. Вам предстоит пройти через две опасные локации: лес и горы.
В лесу есть две возможные тропы - тропа A и тропа B. Вы можете выбрать одну из них случайным образом. Вероятность выбрать тропу A равна 0.4, а тропу B - 0.6. После выбора тропы вы перемещаетесь в горы. Вероятность успешно пройти горы, если вы выбрали тропу A, составляет 0.3. Если же вы выбрали тропу B, вероятность успешного прохождения гор увеличивается до 0.7.


#### Вопрос:
Какова вероятность успешно пройти оба этапа путешествия (попасть в страну X) и как она зависит от выбора тропы в лесу?

#### Решение:
```
P(A) = 0.4
P(B) = 0.6
P(Горы|A) = 0.3
P(Горы|B) = 0.7

P(Успех|A) = P(A) * P(Горы|A) = 0.4*0.3 = 0.12
P(Успех|B) = P(B) * P(Горы|B) = 0.6*0.7 = 0.42

P(Успех) = P(Успех|A)+P(Успех|B) = 0.12+0.42 = 0.54

Общая вероятность успешно завершить путеществие = 54%.  
```

Как успех зависит от выбора тропы в лесу? Не очень понимаю cути задания, но наверное как-то так.
$$
P(A∣Успех)= \frac{P(Успех∣A)⋅P(A)}{P(Успех)} = \frac{0.12}{0.54}=0.222\\
P(B∣Успех)= \frac{P(Успех∣B)⋅P(B)}{P(Успех)} = \frac{0.42}{0.54}=0.778
$$

#### Ответ
* Общая вероятность успешно завершить путеществие = 54%.
* Вероятность завершить путешествие если в лесу была выбрана тропа A = 22.2%
* Вероятность завершить путешествие если в лесу была выбрана тропа B = 77.8%

### 2 Лабиринт сокровищ
#### Задача:
Вы оказались в лабиринте, который славится своими сокровищами. Чтобы добраться до сокровищ, вам нужно пройти через три замка: замок A, замок B и замок C. Каждый замок защищен различными ловушками и подходы к каждому из них различны.

Замок A: Первый замок защищен огненными ловушками. Вероятность успешно пройти замок A составляет 0.6.
Замок B: После успешного прохождения замка A, вы переходите к замку B. Замок B защищен сложными головоломками. Если вы успешно прошли замок A, вероятность успешно пройти замок B составляет 0.4. Однако, если вы не смогли пройти замок A, то вероятность успешно пройти замок B уменьшается до 0.2.
Замок C: После прохождения замка B, вам предстоит столкнуться с последней преградой - замком C. Замок C охраняется могучим воином. Если вы успешно прошли замок B, вероятность успешно пройти замок C составляет 0.7. Однако, если вы не смогли пройти замок B, то вероятность успешно пройти замок C уменьшается до 0.3.

#### Вопрос:
Какова вероятность добраться до сокровища (пройти все три замка) и как она зависит от вашего успешного прохождения предыдущих замков?

#### Решение:
```
P(A) = 0.6
P(¬A) = 1 - P(A) = 1-0.6=0.4

P(B|A) = 0.4
P(B|¬A) = 0.2

P(C|B) = 0.7
P(C|¬B) = 0.3

Формулировка вопроса путает. Но предположим, что нас спрашивали, о том какова вероятность добраться до сокровища, если все три замка УСПЕШНО пройдены. 
P(Успех) = P(A) * P(B|A) * P(C|B)= 0.6*0.4*0.7 = 0.168

Но, что бы добраться до необязательно успешно проходить замки. Попробуем найти вероятность добраться до сокровища с учетом всех возможных путей.
Вероятность пройти замок B:
P(B) = P(B|A)*P(A)+P(B|¬A)*P(¬A) = 0.4*0.6 + 0.2*0.4 = 0.32
Вероятность не пройти замок B:
P(¬B) = 1 - P(B) = 1-0.32 = 0.68

Вероятность пройти замок C:
P(C) = P(C|B)*P(B)+P(C|¬B)*P(¬B) = 0.7*0.32 + 0.3*0.68 = 0.428
Значит полная вероятность пройти замок С и соответственно добраться до сокровища составляет 42.8%.
```
Как вероятность пройти замок С зависит от прохождения замков А и B?  
**Если замок А успешно пройден**
$$
P(A∣C)= \frac{P(C∣A)⋅P(A)}{P(C)}\\
$$
где $P(C∣A)=P(C∣B)*P(B∣A)+P(C∣\neg B)*P(\neg B∣A)$
$$
P(C∣A)=0.7*0.4+0.3*0.6=0.46\\
P(A∣C)=\frac{0.46*0.6}{0.428}=0.645
$$

**Если замок B успешно пройден**
$$
P(B∣C)= \frac{P(C∣B)⋅P(B)}{P(C)}\\
P(B∣C)=\frac{0.7*0.32}{0.428}=0.523
$$

#### Ответ
* Вероятность успешно пройти все три замка 16.8%.
* Полная вероятность пройти замок С 42.8%
* Вероятность пройти замок C, если успешно пройден замок А 64.5%
* Вероятность пройти замок C, если успешно пройден замок B 52.3%

### 3 Кража в ювелирном магазине
#### Задача:
В ювелирном магазине работают два охранника: Алексей и Борис. За последние несколько месяцев в магазине произошла кража драгоценностей. Известно, что Алексей надежный охранник и, когда происходила кража, он был на рабочем месте с вероятностью 0.98. Борис менее ответственный охранник, и он был на рабочем месте с вероятностью 0.85.

Теперь, владелец магазина установил систему видеонаблюдения, которая фиксирует моменты, когда охранники присутствуют на рабочем месте. Согласно системе видеонаблюдения, когда произошла кража, Алексей был на рабочем месте с вероятностью 0.95, а Борис - с вероятностью 0.2.

#### Вопрос:
Если кража все же произошла, какова вероятность того, что она была совершена Алексеем?  

#### Решение
```
P(A на месте)=0.98 — вероятность, что Алексей был на рабочем месте.
P(B на месте)=0.85 — вероятность, что Борис был на рабочем месте.

P(Видео∣A)=0.95 — вероятность, что система зафиксировала Алексея, если он был на месте.
P(Видео∣B)=0.2 — вероятность, что система зафиксировала Бориса, если он был на месте.

Судя по постановке вопроса мы считаем, что подозреваем мы только охранников. Значит априорная вероятность того что Алексей совершил кражу P(A)=P(B)=50%.  
Вероятно охранники не знают о системе видео наблюдения, и следит эта система только за охранниками. Значит доказать факт кражи конкретным охранником можно если кража попала не видео.
```

$$
P(A∣Видео)= \frac{P(Видео∣A)*P(A на месте)*P(A)}{P(Видео)}\\
P(Видео)=P(Видео∣A)*P(A на месте)*P(A)+P(Видео∣B)*P(B на месте)*P(B)=0.98*0.95*0.5+0.85*0.2*0.5=0.5505\\
P(A∣Видео)=\frac{0.98*0.95*0.5}{0.5505}=0.845
$$

#### Ответ
Вероятность того, что Алексей совершил кражу составляет 84.5%.

### 4 Заболевание и результаты тестирования
#### Задача:
Представим, что у определенной популяции есть редкое заболевание, которое встречается у 1% людей. Существует медицинский тест для определения этого заболевания. Тест имеет высокую точность: вероятность правильно определить наличие заболевания у больного составляет 0.95, а вероятность правильно определить отсутствие заболевания у здорового составляет 0.90.

#### Вопрос:
Если у человека положительный тест на заболевание, какова вероятность того, что он действительно болен этим заболеванием?

#### Решение
```
P(A) = 0.01 - заболевание встречается у 1% людей
P(B|A) = 0.95 - если человек болен, то тест точен на 95%
P(¬B|¬A) = 0.9 - вероятность верно определить отсутствие заболевания у здорового человека 90%.
P(B) - общая вероятность положительного теста
```

$$
P(A∣B)= \frac{P(B∣A)*P(A)}{P(B)}\\
P(B)=P(B|A)*P(A)+P(B|\neg A)*P(\neg A)=P(B|A)*P(A)+(1-P(\neg B|\neg A))*(1-P(A))=0.95*0.01+(1-0.9)*(1-0.01)=0.1085\\
P(A∣B)= \frac{0.95*0.01}{0.1085}=0.087
$$

#### Ответ
Вероятность того что человек болен заболеванием при получении положительного теста составляет 8.7%.

### 5 Студенты и предметы
#### Задача:
В университете есть два курса по математике: Алгебра и Анализ. Студенты изучают оба предмета, и 70% студентов посещают курс Алгебры, а 30% студентов посещают курс Анализа. Вероятность того, что студент с курса Алгебры сдаст экзамен, составляет 80%, а вероятность того, что студент с курса Анализа сдаст экзамен, составляет 70%.

#### Вопрос:
Если студент сдал экзамен по математике. Какова вероятность того, что он изучал курс Алгебры?

#### Решение
```
P(Алгебра)=0.7
P(Анализ)=0.3

P(Экзамен|Алгебра)=0.8
P(Экзамен|Анализ)=0.7
```

$$
P(Алгебра∣Экзамен)= \frac{P(Экзамен∣Алгебра)*P(Алгебра)}{P(Экзамен)}\\
P(Экзамен)=P(Алгебра)*P(Экзамен|Алгебра)+P(Анализ)+P(Экзамен|Анализ)=0.7*0.8+0.3*0.7=0.77\\
P(Алгебра∣Экзамен)= \frac{0.8*0.7}{0.77}=0.7272\\
$$

#### Ответ
Вероятность того что чтудент сдавший экзамен изучал алгебру составляет 72.72%.
